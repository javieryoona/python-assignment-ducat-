```python
string = "hello"
r_string = ""
for char in string:
    r_string = char + r_string
print(r_string)
```

```python
def r_string(string1: str) ->str:
    """
    function:
        to revrse the string using for loop 
    args:
        the input must be a string
    result:
        your reversed string is 
    """
    if not isinstance(string1,str):
        raise TypeError ("input should be a string ")
    r_string = ""
    for char in string1:
        r_string= char + r_string
    return r_string

if __name__ == "__main__":
    try:
        user_input = input("enter your string")

        result = r_string(user_input)

        print(f"now your reverse string is: {result}")
    except TypeError as e:
        print(e)
    except Exception as e:
        print(e)
```
